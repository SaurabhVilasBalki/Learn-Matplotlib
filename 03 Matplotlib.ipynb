# Matplotlib

* Matplotlib is the most popular plotting library for Python.
* It gives you control over every aspect of a figure.
* It was designed to have a similar feel to Matlab's graphical plotting.

https://matplotlib.org/stable/users/getting_started/

https://matplotlib.org/stable/gallery/index.html

```python
import matplotlib.pyplot as plt
```

```python
# to see plots inside jupyter notebook
%matplotlib inline
```

### Object-oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [left, bottom, width, height]
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y) # plot using object
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_title('Axes 1')
axes2.set_title('Axes 2')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.2,0.8,0.5])
axes1.set_title('Axes 1')
axes2.set_title('Axes 2')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Axes 1 : x vs y')

axes2.plot(y,x)
axes2.set_title('Axes 2 : y vs x')
```

```python
# SUBPLOTS in Object-oriented method

fig, axes = plt.subplots(nrows=2, ncols=3)

plt.tight_layout() # fixes the overlapping
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('x vs y')

axes[1].plot(y,x)
axes[1].set_title('y vs x')
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(3,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(3,2), dpi=50)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,5))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[0].set_title('x vs y')

axes[1].plot(y,x)
axes[1].set_title('y vs x')

plt.tight_layout()
```

### Adding legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc=0)
```

to check for location of legend https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc=(0.1,0.2))
```

### Setting colors, linewidths, linetypes

### color

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#af42ed') # using rgb hex color code
```

### linewidth

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, linewidth=10) # by default it is 1
```

### linetransparency - alpha

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple', lw=3, alpha=0.5)
```

### line style

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, lw=5, linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, lw=5, ls='-.')
```

**ls supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'**

### markers

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, marker='o')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, marker='x')
```

to check for markers supported values https://matplotlib.org/stable/api/markers_api.html

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, marker='P')
```

### markersize

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, marker='x', markersize=15)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, marker='3', markersize=15)
```

### customize markers

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, marker='o', markersize=15,
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='red')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
